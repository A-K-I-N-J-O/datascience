## Basic Map

```python
draw_map((51.5135015, -0.1362392))
```

## Adding MapPoints

```python
points = [MapPoint((51.5135015, -0.1358392)), MapPoint((51.5137, -0.1358392)), MapPoint((51.5132, -0.138)), MapPoint((51.5143, -0.135))]
draw_map((51.5135020, -0.1362392), points=points)
```

```python
region = MapRegion(geo_path='../data/us-states.json')
draw_map((40, -90), regions=[region], zoom_start=4)
```
